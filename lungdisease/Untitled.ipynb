```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, zscore
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load your CSV file
df = pd.read_csv('GSE150910_gene-level_count_file.csv', index_col=0)

# Check data structure
print("Data shape:", df.shape)
print("First few rows:")
print(df.head())

# Your data appears to be genes x samples, so we'll transpose if needed
# If samples are in columns and genes in rows, transpose:
if df.shape[0] < df.shape[1]:  # More columns than rows suggests genes are rows
    df = df.transpose()

print("After transpose - Data shape:", df.shape)
```

```python
# For GSE150910 (COVID-19 dataset), you need to assign labels based on your sample information
# This is an example - adjust based on your actual sample groups
# You'll need to check the GEO page or metadata to determine which samples are COVID vs Normal

# Example assignment (adjust numbers based on your actual data):
n_samples = df.shape[0]
# Assuming first half are COVID, second half are Normal (check your metadata!)
covid_samples = n_samples // 2
normal_samples = n_samples - covid_samples

labels = ['COVID'] * covid_samples + ['Normal'] * normal_samples
df['label'] = labels

print("Sample distribution:")
print(df['label'].value_counts())
```

```python
# Remove label column for gene analysis
gene_data = df.drop(columns=['label'])

# Perform t-test for each gene
results = []
covid_group = df[df['label'] == 'COVID'].drop(columns=['label'])
normal_group = df[df['label'] == 'Normal'].drop(columns=['label'])

for gene in gene_data.columns:
    try:
        stat, p = ttest_ind(covid_group[gene], normal_group[gene])
        fold_change = covid_group[gene].mean() - normal_group[gene].mean()
        results.append((gene, stat, p, fold_change))
    except:
        continue

# Create results dataframe
result_df = pd.DataFrame(results, columns=['Gene', 'T-statistic', 'P-value', 'FoldChange'])
result_df = result_df.sort_values('P-value')

print("Top 10 differentially expressed genes:")
print(result_df.head(10))
```

```python
# Select top significant gene
top_gene = result_df.iloc[0]['Gene']
print(f"Analyzing gene: {top_gene}")

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='label', y=top_gene, data=df)
plt.title(f'Expression of {top_gene} by Condition')
plt.ylabel("Expression Level")
plt.show()

# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='label', y=top_gene, data=df)
plt.title(f'Expression of {top_gene} by Condition')
plt.ylabel("Expression Level")
plt.show()

# Strip plot
plt.figure(figsize=(8, 6))
sns.stripplot(x='label', y=top_gene, data=df, jitter=True)
plt.title(f'Individual Expression of {top_gene}')
plt.ylabel("Expression Level")
plt.show()
```

```python
# Split data by group
covid_vals = df[df['label'] == 'COVID'][top_gene].values
normal_vals = df[df['label'] == 'Normal'][top_gene].values

plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(covid_vals)), covid_vals, label='COVID', marker='o')
plt.plot(np.arange(len(normal_vals)), normal_vals, label='Normal', marker='x')
plt.title(f"Expression of {top_gene} Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Expression Level")
plt.legend()
plt.show()
```

```python
# Statistical test results
from scipy.stats import ttest_ind

covid_vals = df[df['label'] == 'COVID'][top_gene]
normal_vals = df[df['label'] == 'Normal'][top_gene]

t_stat, p_value = ttest_ind(covid_vals, normal_vals, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3e}")

# Effect size (Cohen's d)
def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    s1, s2 = group1.std(), group2.std()
    pooled_sd = np.sqrt(((n1 - 1) * s1**2 + (n2 - 1) * s2**2) / (n1 + n2 - 2))
    return (group1.mean() - group2.mean()) / pooled_sd

d = cohens_d(covid_vals, normal_vals)
print(f"Cohen's d: {d:.2f}")
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x=top_gene, hue='label', kde=True, element="step", stat="density")
plt.title(f"Histogram of {top_gene} Expression")
plt.show()

# CDF
def plot_cdf(values, label):
    sorted_vals = np.sort(values)
    y = np.arange(len(sorted_vals)) / float(len(sorted_vals))
    plt.plot(sorted_vals, y, label=label)

plt.figure(figsize=(10, 6))
plot_cdf(covid_vals, "COVID")
plot_cdf(normal_vals, "Normal")
plt.title(f"CDF of {top_gene} Expression")
plt.xlabel("Expression Level")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PCA with top genes
top_genes = result_df['Gene'].head(4).tolist()
X = df[top_genes].values
y = df['label'].values

# Normalize
X_scaled = StandardScaler().fit_transform(X)

# PCA
pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)

# Plot
pca_df = pd.DataFrame(data=components, columns=['PC1', 'PC2'])
pca_df['label'] = y

plt.figure(figsize=(10, 8))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='label')
plt.title('PCA of Gene Expression')
plt.show()
```

```python
# Gene correlation heatmap
top_genes = result_df['Gene'].head(6).tolist()
corr = df[top_genes].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Gene Expression Correlation")
plt.show()

# Clustered heatmap
plt.figure(figsize=(12, 8))
sns.clustermap(df[top_genes], cmap="vlag", standard_scale=1)
plt.title("Clustered Heatmap of Gene Expression")
plt.show()
```

```python
# Create volcano plot
result_df['-log10(P)'] = -np.log10(result_df['P-value'])

plt.figure(figsize=(12, 8))
sns.scatterplot(data=result_df, x='FoldChange', y='-log10(P)')
plt.axhline(-np.log10(0.05), color='red', linestyle='--')  # p=0.05 cutoff
plt.axvline(1, color='green', linestyle='--')  # fold change threshold
plt.axvline(-1, color='green', linestyle='--')
plt.title("Volcano Plot")
plt.xlabel("Mean Expression Difference (COVID - Normal)")
plt.ylabel("-log10(P-value)")
plt.show()
```

```python
# Build classifier with top genes
top_genes = result_df['Gene'].head(50).tolist()
X = df[top_genes]
y = df['label']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
pred = clf.predict(X_test)

print("Classifier Accuracy:", accuracy_score(y_test, pred))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Feature selection (e.g., top 500 genes)
top_genes = result_df['Gene'].head(500).tolist()
X = df[top_genes]
y = df['label']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# Hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
svm = GridSearchCV(SVC(), param_grid, cv=5)
svm.fit(X_train, y_train)

print("Best SVM Accuracy:", svm.best_score_)
```

```python
import xgboost as xgb

# Feature selection
X = df[top_genes]
y = df['label'].astype('category').cat.codes  # Convert labels to integers

# Train XGBoost
model = xgb.XGBClassifier(
    objective='multi:softmax',  # For multi-class
    n_estimators=200,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8
)
model.fit(X_train, y_train)
print("XGBoost Accuracy:", model.score(X_test, y_test))
```
